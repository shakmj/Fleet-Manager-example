```python
#second to run
import pandas as pd
file_path = "carData.xlsx";
#file = pd.ExcelFile(file_path)
#file.sheet_names
#third to run
cars_df = pd.read_excel(file_path ,sheet_name="car")
pwr_trains_df = pd.read_excel(file_path ,sheet_name="pwr_train")
owners_df = pd.read_excel(file_path ,sheet_name="owner")
others_df = pd.read_excel(file_path ,sheet_name="others")
```

```python
# 4th to run
# create vehicle objects:
# hint: to use the class
# veh1 = Vehicle_fleet( "registeration date" , "brand_model", "chassis_number","next check date");

#create lists for objects
veh_fleet_list  = [];  # main object list
veh_owner_list  = [];  # owner list
pwr_train_list  = []; # power train list
#other_prp_list  = [];   no need as it will be included as composition in the main object.

# sublist for the date from the Excel file.
# columns from the cars itself table
brand_model        = cars_df["model_brand"];
registeration_date = cars_df["registertion_date"];
chassis_number     = cars_df["chassis number"];
next_check_date    = cars_df["next_check_date"];
# columns from Owner table
owners_names       = owners_df["name"];
plates_numbers     = owners_df["plate_number"];
# columns from power train table
engines_models     = pwr_trains_df["engine_model"];
engines_numbers    = pwr_trains_df["engine_number"];
gearboxs_type      = pwr_trains_df["gearbox_type"];
total_km           = pwr_trains_df["total_km"];
# columns from other properties table
ecu_version        = others_df ["ecu_version"];
coolant            = others_df ["coolant"];
color              = others_df ["color"];

## for loops to create the objects and relations:
for i, i_l in enumerate(brand_model):
        veh_fleet_list.append( Vehicle_fleet( registeration_date[i] , brand_model[i], chassis_number[i],next_check_date[i]) )
for i, i_l in enumerate(owners_names):
        veh_owner_list.append( owners( owners_names[i] , plates_numbers[i]) ); 
        veh_fleet_list[i].set_owner( veh_owner_list[i].Owner_name, veh_owner_list[i].vehicle_plate_number);    # association or? 
for i, i_l in enumerate(engines_models):
        pwr_train_list.append( pwr_train_prp( engines_models[i] ,engines_numbers[i], total_km[i] , gearboxs_type[i])  );
        veh_fleet_list[i].set_power_train(pwr_train_list[i]);                                           #  aggregation!
for i , i_l in enumerate(ecu_version):
        veh_fleet_list[i].set_other_prp(coolant[i] , color[i] , ecu_version[i])

print(brand_model[3]);
```
